```python
#lab task 03
from collections import deque
class WaterJug():
    def __init__(self,capacity_x,capacity_y,goal):
        self.capacity_x=capacity_x
        self.capacity_y=capacity_y
        self.goal=goal
        self.visited =set()
    def dfs(self,x,y):
        if(x,y) in self.visited:
            return None
        self.visited.add((x,y))
        
        print(f"jug1: {x},jug2: {y}")
        if x==self.goal or y==self.goal:
            return True
        if self.dfs(self.capacity_x,y):
            return True
        if self.dfs(x,self.capacity_y):
            return True
        if self.dfs(0, y):
            return True
        
        if self.dfs(x, 0):
            return True
        
        if x + y <= self.capacity_y:
            if self.dfs(0, x + y):
                return True
        else:
            if self.dfs(x - (self.capacity_y - y), self.capacity_y):
                return True
            
        if x + y <= self.capacity_x:
            if self.dfs(x + y, 0):
                return True
        else:
            if self.dfs(self.capacity_x, y - (self.capacity_x - x)):
                return True
            
        return False

    def solve(self):
        return self.dfs(0, 0)
    
def water_jug(capacity_x ,capacity_y,goal):
        result= WaterJug(capacity_x ,capacity_y,goal)
        if not result.solve():
            print("result not found")
        else:
            print("result found")
            
            
            
water_jug(5,9,6)
            
        
    
            
        
        
        
    
    
```

```python
#lab task 4

class NQueen():
    def __init__(self,n):
        self.n=n
        self.board=[-1]*n
        self.result=[]
        
    def is_safe(self,row,col):
        for r in range(row):
            c=self.board[r]
            if c== col or abs (r-row) == abs(c - col):
                return False
        return True
    
    def solve(self, row=0):
        if row == self.n:
            self.result.append(self.board[:])
            return

        for col in range(self.n):
            if self.is_safe(row, col):
                self.board[row] = col
                self.solve(row + 1)
                self.board[row] = -1 

    def print_board(self, solution):
        for row in solution:
            board_row = ['.' for _ in range(self.n)]
            board_row[row] = 'Queen'
            print(' '.join(board_row))
        print()

    def get_solutions(self):
        self.solve()
        for solution in self.result:
            self.print_board(solution)
            
def n_queens_problem(n):
    solver = NQueen(n)
    solver.get_solutions()
    
n_queens_problem(6)
    

            
            
            
        


        
    
```
