```python
y = df['SalePrice']
X = df[['GarageCars', 'GrLivArea', 'YearBuilt', 'Foundation', 'GarageArea', 'YearRemodAdd', 'FullBath', 'GarageYrBlt']]
 
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
X_encoded = pd.get_dummies(X, drop_first=True)


X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_encoded_imputed = imputer.fit_transform(X_encoded)
X_encoded_imputed_df = pd.DataFrame(X_encoded_imputed, columns=X_encoded.columns)
X_train, X_test, y_train, y_test = train_test_split(X_encoded_imputed_df, y, test_size=0.25, random_state=42)

rf = RandomForestRegressor(n_estimators=200, max_depth=10, random_state=42)
rf.fit(X_train, y_train)


prediction = rf.predict(X_test)
prediction
```

```python
scores = cross_val_score(rf, X_encoded_imputed_df, y, cv=5)
print(f"Cross-validated scores: {scores}")
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, prediction)
r2
```
